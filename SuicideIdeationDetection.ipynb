### Capstone I
**Goal:** The goal of this project is to identify better ways to detect people at risk for suicide. We will be training several classification models (Naive Bayes, Logistic Regression, Decision Trees, SVC) to detect suicide ideation from social media posts and comparing their performances. Then, we will further scrutinize the models to identify the features and parameters which most effectively predict whether a user is at risk for suicide.

**Data Problem**: The data task is to train and tune multiple binary text classification models that group social media posts into suicidal vs non-suicidal classes. Further scrutinizing the most effective features used by the best model will indicate measures to more effectively detect suicide ideation.

#### Data Setup

```python
#only keeping the text and class columns
suicide.drop("Unnamed: 0", axis = 1, inplace = True)

#renaming "class" column as "suicide"
suicide.rename({"class" : "suicide"}, axis = 1, inplace = True)

#numerically representing the contents of "suicide" column
suicide.replace({"suicide" : 1, "non-suicide" : 0}, inplace = True)
suicide.head()
```

```python
#train test split
X = suicide['text']
y = suicide['suicide']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
#sample 
X_train.iloc[0]
```

#### Text Preprocessing

```python
#tokenization, stop word removal, punctuation removal, lower casing, and stemming
def preprocess_text(text):
    tokenizer = TweetTokenizer(preserve_case = False)
    swords = stopwords.words('english')
    punc = list(string.punctuation)
    stemmer = PorterStemmer()
    
    proprocessed_text = []
    for entry in text:
        tokens = tokenizer.tokenize(entry)
        proprocessed_text.append([stemmer.stem(word) for word in tokens if word not in swords and word not in punc])
    return proprocessed_text
```

```python
#preprocessing
pX_train = preprocess_text(X_train)
X_train = [' '.join(x) for x in pX_train]

pX_test = preprocess_text(X_test)
X_test = [' '.join(x) for x in pX_test]
```

```python
#checking output 
X_train[0]
```

#### Model Training

Holdout cross validation is implemented; models will be trained on the training set and cross validated with the test set.

RandomizedSearchCV is used to find the optimal number of features for model training.

We will be evaluating the models with the accuracy score metric because we have a balanced dataset. Accuracy measures the proportion of correctly predicted observations out of total observations and is calculated as (True Positive count + True Negative count) / (Total count), or

$$
\frac{TP + TN}{TP + TN + FP + FN}
$$

**Naïve Bayes** 

```python
#naive bayes model with tfidf vectorizer 
nb = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('bayes', MultinomialNB())
])
```

```python
#nb params
nb_params = {'bayes__alpha' : [1, 0.1, 0.01, 0.001, 0]}
```

```python
#grid search 
nb_grid = RandomizedSearchCV(estimator = nb,
                            param_distributions = nb_params)

start = time.time()
nb_grid.fit(X_train, y_train)
stop = time.time()

#stats
nb_time = stop - start
nb_params = nb_grid.best_params_
nb_acc = nb_grid.score(X_test, y_test)
nb_rec = recall_score(y_test, nb_grid.predict(X_test))
nb_f1 = f1_score(y_test, nb_grid.predict(X_test))
```

```python
#best params
nb_params
```

**Logistic Regression**

```python
#logreg + tfidf 
lgr = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('lgr', LogisticRegression())
])
```

```python
#lgr params
lgr_params = {'lgr__penalty' : ['l1', 'l2', 'none'],
             'lgr__fit_intercept' : [True, False]}
```

```python
#gridsearch 
lgr_grid = RandomizedSearchCV(estimator = lgr,
                   param_distributions = lgr_params)


start = time.time()
lgr_grid.fit(X_train, y_train)
stop = time.time()

#stats
lgr_time = stop - start
lgr_params = lgr_grid.best_params_
lgr_acc = lgr_grid.score(X_test, y_test)
lgr_rec = recall_score(y_test, lgr_grid.predict(X_test))
lgr_f1 = f1_score(y_test, lgr_grid.predict(X_test))
```

```python
#best params
lgr_params
```

**Decision Tree**

```python
#decision tree + tfidf 
tree = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('tree', DecisionTreeClassifier())
])
```

```python
#tree params
tree_params = {'tree__criterion' : ['gini', 'entropoy'],
              'tree__max_depth' : [None, 1, 2, 3, 4],
              'tree__min_samples_split' : [1, 2, 3, 4]}
```

```python
#gridsearch
tree_grid = RandomizedSearchCV(estimator = tree,
                   param_distributions = tree_params)

start = time.time()
tree_grid.fit(X_train, y_train)
stop = time.time()

#stats
tree_time = stop - start
tree_params = tree_grid.best_params_
tree_acc = tree_grid.score(X_test, y_test)
tree_rec = recall_score(y_test, tree_grid.predict(X_test))
tree_f1 = f1_score(y_test, tree_grid.predict(X_test))
```

```python
#best params
tree_params
```

**SVC**

```python
#svc + tfidf 
svc = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('svc', SVC(kernel = 'linear'))
])
```

```python
#params
svc_params = {'svc__decision_function_shape' : ['ovo', 'ovr']}
```

```python
#gridsearch
svc_grid = RandomizedSearchCV(estimator = svc,
                   param_distributions = svc_params)

start = time.time()
svc_grid.fit(X_train, y_train)
stop = time.time()

#stats
svc_time = stop - start
svc_params = svc_grid.best_params_
svc_acc = svc_grid.score(X_test, y_test)
svc_rec = recall_score(y_test, svc_grid.predict(X_test))
svc_f1 = f1_score(y_test, svc_grid.predict(X_test))
```

```python
#best params
svc_params
```

**Results**

Best accuracy: **SVC** with a score of 0.924

Best recall: **Naive Bayes** with a score of 0.955

Best f1: **SVC** with a score of 0.924
